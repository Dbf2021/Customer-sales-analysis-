```python
import requests  # Imports the requests library, allowing us to make HTTP requests to APIs

# Make a GET request to the specified API endpoint
response = requests.get("https://jsonplaceholder.typicode.com/todos")

# Check if the request was successful
if response.status_code == 200:
    # If successful, parse and print the JSON response data
    print(response.json())
else:
    # If the request was unsuccessful, print an error message
    print('Failed to retrieve data')
```

Convert json data from api response into pandas dataframe

```python
import pandas as pd  # Imports the pandas library for data manipulation and analysis

# URL of the API endpoint that provides sample data
url = 'https://jsonplaceholder.typicode.com/todos'

# Sends a GET request to the API to retrieve the data
response = requests.get(url)

# Checks if the request was successful
if response.status_code == 200:
    # Parses the JSON response data
    data = response.json()

    # Converts the JSON data into a pandas DataFrame for easy data manipulation and analysis
    df = pd.DataFrame(data)

    # Prints the DataFrame, which displays the data in a tabular format
    print(df)
else:
    # If the request failed, it prints an error message
    print('Failed to retrieve data')
```

Retrieve data from weather data API

```python
import requests
import pandas as pd
#define api key
api_key = '9102eb3de803cdb3b1d84b4db716913d' #replace it with your actual api key
#list of countries to retrieve weather data from
# List of cities for which we want to retrieve weather data
cities = ['Kenya', 'London', 'New York', 'Tokyo', 'Sydney', 'Delhi','Uganda','Nigeria','Tanzania']  # Add more cities as needed
#initialize an empty list to store weather data for each city
weather_data_list = []
#loop through each city and retrieve the weather data
for city in cities:
  #define the api endpoint for each city
   url = f'https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric'
   #make a get request  to the api
   response = requests.get(url)
   #check if the request was successfull
   if response.status_code == 200:
    #parse the json response
    data = response.json()
    #extarct the relevat weather data for the city
    weather_data= {
            'City': data['name'],
            'Temperature (°C)': data['main']['temp'],
            'Humidity (%)': data['main']['humidity'],
            'Weather Description': data['weather'][0]['description'],
            'Wind Speed (m/s)': data['wind']['speed'],
    }

    #append the citys weather data to list
    weather_data_list.append(weather_data)

   else:
      print('unable to retrieve data')
   #convert the list of weather data dictionaries to a dataframe
   df_weather_all_cities = pd.DataFrame(weather_data_list)
   #display the dataframe
   print(df_weather_all_cities)
```
